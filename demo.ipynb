## Basic Plot Demo
This includes just a simple straightforward bar plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

top20_deathtoll = pd.read_csv('top20_deathtoll.csv')
plt.barh(top20_deathtoll['Country_Other'],
         top20_deathtoll['Total_Deaths'])
plt.show()
```

Now we can refine this by progressively applying different reductions.

First we build out a "figure" and then plot again a basic bar-plot. 
However, this time we now reduce the size of the bars and change the color.

```python
# First we build out a "figure" and then plot again a basic bar-plot. 
# However, this time we now reduce the size of the bars and change the color.

fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'],
        height=0.45, color='#af0b1e')  # Reduce the thickness of bars

plt.show()
```

Now we remove the frame from the plot.

```python
fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'],
        height=0.45, color='#af0b1e')  

#======================================================
# NEW BELOW:
#======================================================

# Remove the frame around the plot
for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False)

plt.show()
```

And we reduce and move the axis labels.

Along with removing all the tick marks.

```python
fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'],
        height=0.45, color='#af0b1e')  

# Remove the frame around the plot
for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False)


#======================================================
# NEW BELOW:
#======================================================

ax.set_xticks([0, 150000, 300000])  # Set specific x-axis ticks
ax.xaxis.tick_top()  # Move x-axis labels to the top
ax.tick_params(top=False, left=False)  # Remove tick marks


plt.show()

```

Finally, we lighten the labels and add a vertical line to help guide the eye.

NOTE: Vertical line was NOT in the slides.

The original code did some additional things like left-align the text (manually)

```python
fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'],
        height=0.45, color='#af0b1e')  

# Remove the frame around the plot
for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False)

ax.set_xticks([0, 150000, 300000])  # Set specific x-axis ticks
ax.xaxis.tick_top()  # Move x-axis labels to the top
ax.tick_params(top=False, left=False)  # Remove tick marks


#======================================================
# NEW BELOW:
#======================================================
ax.tick_params(axis='x', colors='grey')  # Color x-axis labels grey
ax.axvline(x=150000, ymin=0.045, c='grey', alpha=0.5)


plt.show()
```

Now let's use the gestalt principle of similarity to color North America and South America differently.

This requires specifying a list of specific colors for each datapoint.

```python

fig, ax = plt.subplots(figsize=(4.5, 6))
bars = ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'],
        height=0.45, color='#af0b1e')  

# Remove the frame around the plot
for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False)

ax.set_xticks([0, 150000, 300000])  # Set specific x-axis ticks
ax.xaxis.tick_top()  # Move x-axis labels to the top
ax.tick_params(top=False, left=False)  # Remove tick marks

ax.tick_params(axis='x', colors='grey')  # Color x-axis labels grey
ax.axvline(x=150000, ymin=0.045, c='grey', alpha=0.5)

#======================================================
# NEW BELOW:
#======================================================
for bar, value in zip(bars, top20_deathtoll['Country_Other']):
    if value == 'USA' or value == 'Mexico':
        bar.set_color('green')
    if value in ('Argentina', "Colombia", "Peru", "Chile"):
        bar.set_color('blue')


#======================================================



plt.show()
```
